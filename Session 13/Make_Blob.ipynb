```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs 
from sklearn.linear_model import LogisticRegression
import seaborn as sb
```

```python
X, y = make_blobs(centers=3, random_state=40)
```

```python
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.classes_
```

```python
colors = ["Red", "Yellow", "Green", "Orange", "Black"]
```

```python
y_col = [colors[item] for item in y]
```

```python
plt.scatter(X[:,0], X[:,1], c=y_col)
```

```python
sb.scatterplot(X[:,0], X[:,1], hue=y)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_col)

for [m1, m2], c in zip(model.coef_, model.intercept_):
    x1 = np.linspace(-10, 10, 2)
    x2 = (-m1/m2)*x1 -(c/m2) 
    
    plt.plot(x1, x2)
    
plt.xlim(-10, 10)
plt.ylim(-10,10)
    
```
