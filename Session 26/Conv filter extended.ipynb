```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img
```

```python
image = np.array(load_img("../photos/mon.jpeg", target_size=(400, 600)))
gray = np.mean(image, axis=2)
```

```python
plt.imshow(gray, cmap="gray")
```

```python
fil = np.array([[1, 0, -1],
               [1, 0, -1],
               [1, 0, -1]])
```

```python
padded_img = np.pad(gray, 1)
```

```python
padded_img.shape
```

```python
result = np.empty_like(gray)
```

```python
def apply_filter(fil, gray):
    
    row, col = 100, 100
    
    for row in range(result.shape[0]):
        for col in range(result.shape[1]):
            cut = padded_img[row:row+3, col:col+3]    
            result[row][col]  = np.abs(np.mean(fil * cut))
        
    return result
```

```python
plt.imshow(apply_filter(fil, gray), cmap="gray")
```

```python
fil2 = np.array([[1, 1, 1],
               [0, 0, 0],
               [-1, -1, -1]])

plt.imshow(apply_filter(fil2, gray), cmap="gray")
```

```python
fil3 = np.array([[1, 1, 1],
               [1, 1, 1],
               [1, 1, 1]])

plt.imshow(apply_filter(fil3, gray), cmap="gray")
```
