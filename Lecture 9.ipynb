```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.ylim(-100, 100)
```

```python
plt.figure()

plt.plot(x, x**2, label="Square")
plt.plot(x, x**3, label="Cube")
plt.xlabel("X- axis")
plt.ylabel("Y- axis")
plt.title("Diagram")
plt.legend()

plt.ylim(-100, 100)

plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.tan(x))
```

```python
x = np.linspace(-10, 10, 100)

plt.figure()
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.tan(x), label="tan")
plt.legend()
plt.ylim(-4, 4)
plt.show()
```

```python
theta = np.linspace(0, 20*np.pi, 1000)
rad = np.linspace(0, 100, 1000)
```

```python
x = rad * np.cos(theta)
y = rad * np.sin(theta)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

```python
theta = np.linspace(0, 40*np.pi, 100)
rad = np.linspace(0, 100, 100)

x = rad * np.cos(theta)
y = rad * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x,y, marker="*")
```
